```python
df2 = pd.DataFrame(Heart,columns=['Cholesterol','FastingBS','Age'])
import statsmodels.formula.api as smf
model = smf.ols(formula='Age ~ Cholesterol + FastingBS', data=df2)
results_formula = model.fit()
results_formula.params
```
